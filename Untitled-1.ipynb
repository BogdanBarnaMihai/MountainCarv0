```python
import numpy as np
import gymnasium as gym
import time 

class q_learning:
    def __init__ (self,env,alpha,gamma,epsilon,numberofepisodes,numberofbins,lowerbounds,upperbounds):
        self.env=env
        self.alpha=alpha
        self.gamma=gamma
        self.epsilon=epsilon
        self.numberofepisodes=numberofepisodes
        self.numberofbins=numberofbins
        self.lowerbounds=lowerbounds
        self.upperbounds=upperbounds

        self.sumrewards=[]
        self.actionnumber=env.action_space.n

        self.qmatrix=np.random.uniform(low=0,high=1,size=(numberofbins[0],numberofbins[1],self.actionnumber))

    def returnindex(self,state):
        position=np.linspace(self.lowerbounds[0],self.upperbounds[0],self.numberofbins[0])
        velocity=np.linspace(self.lowerbounds[1],self.upperbounds[1],self.numberofbins[1])

        positionindex=np.maximum(np.digitize(state[0],position)-1,0)
        velocityindex=np.maximum(np.digitize(state[1],velocity)-1,0)

        return tuple([positionindex,velocityindex])
    
    def select_action(self,state,index):
        if index<500:
            return np.random.choice(self.actionnumber)
        if index>1500:
            self.epsilon*=0.999
        randomaction=np.random.random()
        if randomaction<self.epsilon:
            return np.random.choice(self.actionnumber)
        else:
            return np.random.choice(np.where(self.qmatrix[self.returnindex(state)]==np.max(self.qmatrix[self.returnindex(state)]))[0])
    
    def simulateep(self):
        for indexep in range(self.numberofepisodes):
            if(indexep%10==0):
                print(indexep)
            state,_=self.env.reset()
            state=list(state)
            terminalstate=False
            sumrewards=[]
            while not terminalstate:
                stateindex=self.returnindex(state)
                actiona=self.select_action(state,indexep)
                stateprime,reward,terminalstate,truncated,_=self.env.step(actiona)
                terminalstate=terminalstate or truncated
                if stateprime[0]>0.5:
                    reward+=100
                elif stateprime[0]>0.3:
                    reward+=5
                elif stateprime[0]>0.1:
                    reward+=1
                sumrewards.append(reward)
                stateprime=list(stateprime)
                stateprimeindex=self.returnindex(stateprime)
                qmaxprime=np.max(self.qmatrix[stateprimeindex])
                if not terminalstate:
                    error=reward+self.gamma*qmaxprime-self.qmatrix[stateindex+(actiona,)]
                else:
                    error=reward-self.qmatrix[stateindex+(actiona,)]
                self.qmatrix[stateindex+(actiona,)]+=self.alpha*error
                state=stateprime
            self.sumrewards.append(np.sum(sumrewards))

    def simulatelearnedstrat(self):
        env1=gym.make('MountainCar-v0',render_mode='human')
        state,_=env1.reset() 
        timesteep=1000
        sumrewards=[]
        for i in range(timesteep):
            actiona=np.random.choice(np.where(self.qmatrix[self.returnindex(state)]==np.max(self.qmatrix[self.returnindex(state)]))[0])
            state,reward,terminal,_,_=env1.step(actiona)
            sumrewards.append(reward)
            time.sleep(0.05)
            env1.render()
            if terminal:
                time.sleep(1)
                break
        return (sumrewards,env1)
```

```python
env=gym.make('MountainCar-v0')
state,_=env.reset() 

upperbounds=env.observation_space.high
lowerbounds=env.observation_space.low

numberofbins=[20,20]

alpha = 0.1
gamma = 1
epsilon = 0.2
numberEpisodes = 3500

q1=q_learning(env,alpha,gamma,epsilon,numberEpisodes,numberofbins,lowerbounds,upperbounds)
q1.simulateep()

print(np.mean(q1.sumrewards))
```
